```python
from pydantic import BaseModel , Field
from typing import Optional

class Address(BaseModel):
    street : str
    area : str
    pin : int=Field(ge=10000 , le=999999)   # (min_length=5, max_length=6)// min max for list str 
    state : str

class Person(BaseModel):
    name : str
    age : int
    address : Address
    gender : Optional[str] = None
    salary : Optional[int] = None
    isMarried : Optional[bool] = None #shorthand for Union[X,None]

class Team(BaseModel):
    team : list[str]
    capacity : int

address1 = {
    "street" : "100",
    "area" : "puz",
    "pin" : 10000,
    "state" : "TN"
}

person1 = Person(name="san",age=10,address=address1)
print(person1)
team1 = Team(team=["a","b"],capacity=10)
print(team1)

print(Team.model_json_schema())

```
